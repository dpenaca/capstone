**¿Cómo obtenemos la variable "yearweek"?**

Premisas:
  - El menor valor de "yearweek" en "item_prices.csv" es "201105" y el mayor es "201617"
  - El primer día registrado en "daily_calendar_with_events.csv" es el "d-1" que es un día "sábado" y tiene asignado como "weekday_int" el día "1".  Además indica que es el día 29/01/2011.

Conclusiones:  
Observando la distribución de los datos de "yearweek", "weekday", "weekday_int" y "day" exitentes en "daily_calendar_with_events.csv" y "item_prices.csv", hemos determinado que el cálculo del yearweek no utiliza el estándar que se puede obtener a través del isocalendar, sino que sigue las siguientes reglas:
   - la semana empieza en sábado, que es el día 1 de la semana y el último el viernes, que es el día 7 de la semana
   - la semana 1 de cualquier año empieza a partir del 1er sábado del año, los días anteriores del año pertenecen a la semana 0 del año

Basados en estas premisas incluiremos una columna "yearweek" al dataframe "df_calendario" calculada para adecuarse a las reglas precedentes

```python
# Definición de función para obtener la weekyear dada un fecha de acuerdo a las premisas anteriores

def a_yearweek(fecha):

  # inicializa una variable con el primer día del año de 'fecha'
  primer_dia_any = datetime(fecha.year,1,1)

  # obtiene el día de la semana del primer día del año (0 es lunes, 6 es domingo)
  dia_sem_primer_dia_any = primer_dia_any.weekday()

  #ajuste para adaptar el inicio de semana de lunes a sabado
  ajuste_sabado = (dia_sem_primer_dia_any + 2) % 7

  # determina el número correspondiente a la primera semana del año
  if dia_sem_primer_dia_any != 5:  # si el primer día del año no es sábado
      primera_semana = 0
  else:
    primera_semana = 1

  # calcula la cantidad de días que han transcurrido desde el inicio del año
  diferencia_dias = (fecha - primer_dia_any).days + ajuste_sabado

  # determina la cantidad de semanas desde el inicio del año
  numero_semana = primera_semana + (diferencia_dias  // 7)


  #retorna la respuesta en el formato yearweek
  return fecha.year * 100 + numero_semana

```

```python
df_calendario["yearweek"] = df_calendario['date'].apply(lambda x: a_yearweek(x))
```
